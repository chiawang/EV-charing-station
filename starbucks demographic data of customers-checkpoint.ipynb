<h2>Preparing Data</h2>
We need to do some cleaning on the data before exploring the data

Profile:
<ol>
<li>Delete age = 118 as it doesn't make sense and affect the accuracy of the data</li>
<li>Delete income = NaN to ensure the result of this data are as accurate as possible</li>
<li>Delete gender = 'None'. If the value 'None' coresponds with the age '118' and income 'Nan', this data will be removed.</li>
<li>Change the values in 'id' column to a numberial number</li>
</ol>

```python
profile.dropna(inplace=True) # drops all null values
profile.reset_index(drop=True, inplace=True) # resets all the index of the new records after empty data row has been dropped
profile.isnull().sum() # checks to see if the new clean data set has any null values
```

```python
profile['became_member_on'] = pd.to_datetime(profile['became_member_on'], format='%Y%m%d') # update date format
profile.head() # displays dataframe
```

```python
#create a new label to separate the customer age groups

def separate_data_into_age_group(df):
    bins = [18, 30, 40, 50, 60, 70, 120] # bins for grouped data
    labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+'] # labels of grouped data
    df['age_range'] = pd.cut(df.age, bins, labels = labels,include_lowest = True)
    
    return df.head()


separate_data_into_age_group(profile)
```
