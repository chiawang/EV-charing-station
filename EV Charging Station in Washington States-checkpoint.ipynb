```python
#importing the data back into the notebook
import requests
import time
import pandas as pd
df=pd.read_csv('Electric_Vehicle_Title_and_Registration_Activity.csv', index_col=0)
df.info()
```

<h2>Preparing the data</h2>

```python
#display all columns
#pd.set_option('max_columns', None)
#df.head()
```

```python
# dropping unneed goverment columns
df.drop(['Electric Vehicle Fee Paid'], axis=1, inplace=True)
df.drop(['2015 HB 2778 Exemption Eligibility'], axis=1, inplace=True)
df.drop(['2019 HB 2042 Clean Alternative Fuel Vehicle (CAFV) Eligibility'], axis=1, inplace=True)
df.drop(['Meets 2019 HB 2042 Electric Range Requirement'], axis=1, inplace=True)
df.drop(['Meets 2019 HB 2042 Sale Date Requirement'], axis=1, inplace=True)
df.drop(['Meets 2019 HB 2042 Sale Price/Value Requirement'], axis=1, inplace=True)
df.drop(['2019 HB 2042: Battery Range Requirement'], axis=1, inplace=True)
df.drop(['2019 HB 2042: Purchase Date Requirement'], axis=1, inplace=True)
df.drop(['2019 HB 2042: Sale Price/Value Requirement'], axis=1, inplace=True)
df.drop(['Transportation Electrification Fee Paid'], axis=1, inplace=True)
df.drop(['Hybrid Vehicle Electrification Fee Paid'], axis=1, inplace=True)
df.drop(['2020 Census Tract'], axis=1, inplace=True)
df.drop(['Legislative District'], axis=1, inplace=True)
df.drop(['Electric Utility'], axis=1, inplace=True)

df.info()
```

```python
df[df.duplicated()] #Exact Duplicates
```

```python
# delete duplicates
df.drop_duplicates(inplace=True)

# verifying that duplicates have been removed
len(df[df.duplicated()])
```

```python
# checking if there's null values in the dataset
df.isna().sum()
```

'Sale Date' column seems to be missing quite a bit of information. We have the dta of 'DOL transation date' which is the date when they register the car with the state. In general, the 'DOL transaction date' and the actual vehicle sale are very close (within a few weeks). The 'Sale date' here aren't too meaningful to us and seems to be missing a lot of data and so we are going to drop this colomun.

```python
df.drop(['Sale Date'], axis=1, inplace=True) 
df.drop(['Base MSRP'], axis=1, inplace=True) # we have sale price already and so don't need this one
df.drop(['Transaction Year'], axis=1, inplace=True) # we have DOL transaction date also and so don't need this either
```

```python
df.isna().sum() #checking the null one again after dropping the columns that are no longer needed
```

```python
#dropping null values
df.dropna(subset=['Model'], inplace=True)
df.dropna(subset=['City'], inplace=True)
df.dropna(subset=['County'], inplace=True)
df.dropna(subset=['State of Residence'], inplace=True)
df.dropna(subset=['Postal Code'], inplace=True)
df.isna().sum()
```

<h5> The goal is this project is to focusing on the data analysis in Washington State. Therefore, we are only doing "State of Residence' = WA </h5>

```python
df=df[df['State of Residence']=='WA']
from matplotlib import pyplot as plt
WA_States_EV_Car_Counts = df['County'].value_counts()/len(df['County'])*100 
WA_States_EV_Car_Counts.plot(kind= 'bar', figsize=(17,8)) # plots bar chart
plt.suptitle("Electric Vehicles Counts Per County in Washington State 2023") # title of plot
plt.xlabel('County In Washington State 2023') # label for x axis
plt.ylabel('Percent %'); # label for y axis

df['County'].value_counts()
```
