```python
def EV_car_model_purchased_by_county(df=df, top_n=10, county_list=all_counties):
    
    fig, ax = plt.subplots(figsize=(10, 60), nrows=len(county_list))
    
    for i, county in enumerate(county_list):
        df_county = df[(df['County']==county) & 
                       (df['Transaction Type']=='Original Title')]
        model_count_df = pd.DataFrame(df_county['Model'].value_counts()).reset_index()
        model_count_df.columns=['Model', 'Count']
        sns.barplot(x='Model', y='Count', data=model_count_df.head(top_n), 
                    ax=ax[i])
        ax[i].set_title(f'{county} County (2010-2023)')
    plt.tight_layout();
```

```python
df_cumsum = df_cumsum.set_index('DOL Transaction Date').resample('M').asfreq()
df_cumsum.info()
```

```python
def separate_data_by_county(county, df): 
    df_county = pd.DataFrame(df.loc[:, county])
    df_county.columns=["EV car counts"]
    df_county.head()
    return df_county
```

## Model and Prediction using ARIMA
SARIMAX(Seasonal Auto-Regressive Integrated Moving Average with eXogenous factors) is an extension of the ARIMA class of models. We are using this to make predction to understand how many more EV cars will be on the road in the following county (King County, Snohomish, Pierce) for the next 10 years.


```python
#separating all the county data with its EV car counts
county_information = {}
for county in all_counties:
    county_information[county] = {}
    county_information[county]['df']=separate_data_by_county(county, df_cumsum)
```

```python
def get_prediction(model, df, test_data, county_name, plot=True):    
    #creating a df with the prediction information
    prediction_df = model.get_forecast(steps=len(test_data)*5).conf_int()
    prediction_df.columns = ['Lower Confidence Interval', 'Upper Confidence Interval']
    prediction_df['Predictions'] = model.get_forecast(steps=len(test_data)*5)\
    .predicted_mean
    #plotting
    if plot==True:
        with plt.style.context('seaborn-whitegrid'):
            fig, ax = plt.subplots(figsize=(20, 15))
            sns.lineplot(data=df, ax=ax)
            sns.lineplot(data=prediction_df, x=prediction_df.index, 
                         y='Predictions', color='red', ax=ax, 
                         label='Predicted Data', ls='--')
            ax.fill_between(prediction_df.index, 
                            y1=prediction_df['Lower Confidence Interval'], 
                            y2=prediction_df['Upper Confidence Interval'],
                            color = 'green', alpha=0.3, 
                            label='Confidence Interval')
            ax.set_xlabel('Year')
            ax.set_ylabel('Electric Vehicles on the Road')
            ax.set_title(f'Predicted Electric Vehicle Count for {county_name}')
            plt.show();
    return prediction_df
```

```python
def train_test_split_ts (df, train_size, test_size):
    train_end_idx = int(round(len(df)*train_size,0))
    train_set = df.iloc[0:train_end_idx,:]["EV car counts"]
    test_set = df.iloc[train_end_idx:,:]["EV car counts"]
    return train_set, test_set
```

<h3> King County EV Car Prediction </h3>

```python
#splitting dataset into train and test sets for validation
train_king, test_king = train_test_split_ts(county_information['King']['df'], 0.80, 0.20)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
King_County_Model = SARIMAX(county_information['King']['df'], order=(1,1,0), 
                seasonal_order=(0,1,0,12), enforce_invertibility=False, 
                enforce_stationarity=False).fit()
print(King_County_Model.summary())
```

```python
df_king_preds = get_prediction(King_County_Model, county_information['King']['df'], test_king, 'King', plot=True)
```
